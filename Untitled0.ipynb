```python
import pandas as pd

# 업로드한 파일 이름 그대로 사용
df = pd.read_csv("StudentsPerformance.csv")

# 데이터 미리 보기
df.head()
```

```python
overall_avg = df[['math score', 'reading score', 'writing score']].mean()
print("전체 평균 점수:\n", overall_avg)
```

```python
group_avg = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean().round(2)
print("Group별 평균 점수:\n", group_avg)
```

```python
group_c = df[df['race/ethnicity'] == 'group C']
group_c_avg = group_c.groupby('gender')[['math score', 'reading score', 'writing score']].mean().round(2)
print("Group C의 남자/여자 평균 점수:\n", group_c_avg)
```

```python
male_math_80 = df[(df['gender'] == 'male') & (df['math score'] >= 80)]
print("수학 점수 80점 이상인 남학생 수:", len(male_math_80))
```

```python
completed_prep = df[df['test preparation course'] == 'completed']
completed_prep.head()
```

```python
import numpy as np

# 랜덤 점수 고정 (재현성)
np.random.seed(42)

# 복사본 생성 후 컬럼 추가
completed_prep = completed_prep.copy()
completed_prep['listening score'] = np.random.randint(0, 101, size=len(completed_prep))

completed_prep.head()
```

```python
completed_prep.to_csv("Filtered_student_score.csv", index=False)
print("Filtered_student_score.csv 파일 저장 완료!")
```
