```python
import pandas as pd

# 업로드한 Kaggle 데이터 불러오기
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
# object 타입(범주형) 컬럼만 추출
categorical_columns = df.select_dtypes(include='object').columns

# 각 범주형 컬럼의 고유값 확인
for col in categorical_columns:
    print(f"{col} – {df[col].unique()}")
```

```python
import numpy as np

# 랜덤 데이터 500명 분량 생성
num_samples = 500
random_data = {}

# 각 범주형 컬럼에서 기존 값 중 랜덤하게 선택
for col in categorical_columns:
    random_data[col] = np.random.choice(df[col].unique(), num_samples)

# 수치형 컬럼 (0~100 랜덤 점수)
score_columns = ['math score', 'reading score', 'writing score']
for col in score_columns:
    random_data[col] = np.random.randint(0, 101, size=num_samples)

# 데이터프레임으로 변환
random_df = pd.DataFrame(random_data)
random_df.head()
```

```python
# 두 데이터를 합쳐서 하나의 큰 데이터프레임으로 만듦
combined_df = pd.concat([df, random_df], ignore_index=True)
combined_df.shape  # 행 개수 확인
```

```python
hs_avg = combined_df[combined_df['parental level of education'] == 'high school']['math score'].mean()

ba_avg = combined_df[combined_df['parental level of education'] == "bachelor's degree"]['math score'].mean()

print(f"High School 평균 수학 점수: {hs_avg:.2f}")
print(f"Bachelor's Degree 평균 수학 점수: {ba_avg:.2f}")
```

```python
group_stats = combined_df.groupby('race/ethnicity')[score_columns].agg(['mean', 'std', 'median']).round(2)
group_stats
```

```python
combined_df['average score'] = combined_df[score_columns].mean(axis=1).round(2)
combined_df.head()
```

```python
combined_df.to_csv("New_student_score.csv", index=False)
print("New_student_score.csv 저장 완료!")
```
